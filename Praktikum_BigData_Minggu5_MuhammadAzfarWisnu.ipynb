1. Lakukan teknik data cleaning untuk data yang bernilai NaN pada Dataset 2.

```python
import numpy as np 
import pandas as pd 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data  = pd.read_csv("https://gitlab.com/andreass.bayu/file-directory/-/raw/main/adult.csv")
data.head(10)
```

```python
#menampilkan data kosong
data.isnull().sum().sort_values(ascending=False)
```

```python
#drop / hapus data missing values
data = data.drop(data[data.isnull().any(1)].index, axis=0)
```

```python
data.shape
```

2.Lakukan analisis histogram kolom income pada setiap kategori race, apakah yang dapat Anda simpulkan ?

```python
plt.hist(data["race"], bins=len(data.columns))
plt.ylabel = ("income")
plt.xlabel = ("race")
plt.title("Data income - race")
plt.show()
```

1. Pada setiap race memiliki income yang berbeda
2. Nilai setiap race mempengaruhi perubahan pada grafik historagem
